```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df = pd.DataFrame()
df["temperature"] = [40,41,46,42,45,41,49,56,180,175,53,61,73,87,95,101,107,60]
x = df["temperature"]
```

```
plt.boxplot(x,showmeans=True,sym='b+')
```

**Box Plot Observations:** 1) From the above visual, it is clear that most of the days have higher temperatures as the area of the IQR box is more above the median.
2) There are 2 outliers in the data which fall out of the maximum.  These can be removed.

```
restaurant = pd.DataFrame()
restaurant['age_m'] = [4,5,65,32,64,23,8,34,7,34,11,55,57,33,42,64,24,13,29]
restaurant['age_wo'] = [73,2,13,65,54,47,11,25,29,54,46,37,34,71,69,52,15,44,32]
x=restaurant.age_m
y=restaurant.age_wo
```

```
plt.xlabel("Age of the customers")
plt.ylabel("Total number of customers")
plt.title("Customer age analysis")
plt.hist([x,y],bins=4,rwidth=0.95,label=["men","women"],color=['g','b'])
plt.legend()
```

**3)Line plots, Scatter plots and Subplot**

```
df = pd.DataFrame()
df['rvalues'] = np.random.rand(100)
df['rnormal'] = np.random.normal(100)
df['sum'] = df.rvalues + df.rnormal
df['multiply'] = df.rvalues * df.rnormal
```

```
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.plot(df['sum'],color='b')

plt.subplot(1,2,2)
plt.scatter(x=df['sum'],y=df['multiply'],color='red')
```
